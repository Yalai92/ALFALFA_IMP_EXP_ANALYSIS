```python
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str), format='%Y-%m')
df.set_index('date', inplace=True)

def reindex_and_interpolate(group):
    # Reindexar el grupo para incluir todos los meses
    group = group.reindex(pd.date_range(start=group.index.min(), end=group.index.max(), freq='MS'))

    # Interpolar los valores faltantes
    group['eur_kg'] = group['eur_kg'].interpolate(method='linear')

    # Volver a agregar las columnas 'Año' y 'Mes'
    group['year'] = group.index.year
    group['month'] = group.index.month

    return group

df = df.groupby(['country', 'taric']).apply(reindex_and_interpolate)
df = df.reset_index(drop=True)
```
